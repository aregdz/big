##### **Описание**


  - ```-``` — списание средств (расходная транзакция);
 - ```term_id``` — идентификатор терминала.


 ### Таблица ```gender_train.csv```

## Задание 1

1. Для столбца `tr_type` датафрейма `transactions` выберите произвольные `1000` строк с помощью метода `sample()`.
2. В полученной на предыдущем этапе подвыборке найдите долю транзакций (стобец `tr_description` в датафрейме `tr_types`), в которой содержится подстрока `'POS'` или `'ATM'`.


````python
# Задание 1
# 1. Выборка 1000 случайных строк
sample_tr_types = transactions['tr_type'].sample(1000, random_state=42)

# 2. Подсчет доли транзакций с подстрокой 'POS' или 'ATM'
# Создаем словарь для быстрого поиска описаний
tr_types_dict = dict(zip(tr_types['tr_type'], tr_types['tr_description']))

# Функция для проверки наличия подстрок
def contains_pos_atm(tr_type):
    description = tr_types_dict.get(tr_type, '')
    return 'POS' in description or 'ATM' in description

# Применяем функцию к выборке
matches = sample_tr_types.apply(contains_pos_atm)
proportion = matches.mean()

print(f"Доля транзакций, содержащих 'POS' или 'ATM': {proportion:.4f} ({proportion*100:.2f}%)")
````

## Задание 2


1. Для столбца `tr_type` датафрейма `transactions` посчитайте частоту встречаемости всех типов транзакций `tr_type` в `transactions`.
2. Выведите датафрейм, содержащий топ-10 транзакций, отсортированных по убыванию частоты встречаемости. Датафрейм должен содержать столбцы `tr_type`, `tr_description` и столбец с частотой встречаемости.

````python
# 1. Частота встречаемости типов транзакций
tr_type_counts = transactions['tr_type'].value_counts().reset_index()
tr_type_counts.columns = ['tr_type', 'frequency']

# 2. Топ-10 транзакций по частоте
top_10_tr_types = tr_type_counts.head(10)

# 3. Добавляем описания (проверяем наличие tr_types)
if 'tr_types' in globals() or 'tr_types' in locals():
    # Проверяем, есть ли нужные столбцы в tr_types
    if all(col in tr_types.columns for col in ['tr_type', 'tr_description']):
        top_10_with_desc = top_10_tr_types.merge(tr_types[['tr_type', 'tr_description']],
                                                 on='tr_type',
                                                 how='left')

        print("Топ-10 транзакций по частоте:")
        print(top_10_with_desc[['tr_type', 'tr_description', 'frequency']])
    else:
        print("В tr_types отсутствуют необходимые столбцы 'tr_type' и/или 'tr_description'")
        print("Выводим только типы транзакций:")
        print(top_10_tr_types[['tr_type', 'frequency']])
else:
    print("DataFrame tr_types не найден")
    print("Выводим только типы транзакций:")
    print(top_10_tr_types[['tr_type', 'frequency']])
````

## Задание 3
1. В датафрейме `transactions` найдите клиента с максимальной суммой приходов на карту.
2. В датафрейме `transactions` найдите клиента с максимальной суммой расходов по карте.
3. Найдите модуль разницы для этих клиентов между суммой расходов и суммой приходов.

````python
# 1. Разделим транзакции на приходы (положительные amount) и расходы (отрицательные amount)
# Приходы (пополнения)
incomes = transactions[transactions['amount'] > 0].copy()
# Расходы (списания) - делаем сумму положительной для удобства
expenses = transactions[transactions['amount'] < 0].copy()
expenses['amount'] = expenses['amount'].abs()  # Берем модуль для суммирования расходов

# 2. Суммируем приходы и расходы по каждому клиенту
income_by_customer = incomes.groupby('customer_id')['amount'].sum()
expense_by_customer = expenses.groupby('customer_id')['amount'].sum()

# 3. Находим клиента с максимальной суммой приходов
max_income_customer = income_by_customer.idxmax()
max_income_amount = income_by_customer.max()

# 4. Находим клиента с максимальной суммой расходов
max_expense_customer = expense_by_customer.idxmax()
max_expense_amount = expense_by_customer.max()

# 5. Вычисляем сумму приходов и расходов для каждого из найденных клиентов
# Для клиента с максимальными приходами
income_of_max_income_customer = max_income_amount
expense_of_max_income_customer = expense_by_customer.get(max_income_customer, 0)

# Для клиента с максимальными расходами
income_of_max_expense_customer = income_by_customer.get(max_expense_customer, 0)
expense_of_max_expense_customer = max_expense_amount

# 6. Вычисляем модуль разницы между суммой расходов и суммой приходов для каждого клиента
# diff = |расходы - приходы|
diff_max_income = abs(expense_of_max_income_customer - income_of_max_income_customer)
diff_max_expense = abs(expense_of_max_expense_customer - income_of_max_expense_customer)

# 7. Выводим результаты
print("=" * 60)
print("Клиент с максимальной суммой ПРИХОДОВ на карту:")
print(f"  customer_id: {max_income_customer}")
print(f"  Сумма приходов: {income_of_max_income_customer:,.2f}")
print(f"  Сумма расходов: {expense_of_max_income_customer:,.2f}")
print(f"  Модуль разницы (|расходы - приходы|): {diff_max_income:,.2f}")
print()

print("Клиент с максимальной суммой РАСХОДОВ по карте:")
print(f"  customer_id: {max_expense_customer}")
print(f"  Сумма приходов: {income_of_max_expense_customer:,.2f}")
print(f"  Сумма расходов: {expense_of_max_expense_customer:,.2f}")
print(f"  Модуль разницы (|расходы - приходы|): {diff_max_expense:,.2f}")
print("=" * 60)

# Дополнительно: если это один и тот же клиент
if max_income_customer == max_expense_customer:
    print(f"Примечание: Это один и тот же клиент ({max_income_customer})")
    print(f"Общий баланс (приходы - расходы): {income_of_max_income_customer - expense_of_max_income_customer:,.2f}")
````

## Задание 7

1. Найдите суммы затрат по каждой категории (`mcc`) для мужчин и для женщин.
2. Найдите топ-10 категорий с самыми большими относительными модулями разности в тратах для разных полов (в ответе должны присутствовать описания mcc-кодов).

````python
# Задание 7
# 1. Суммы затрат по категориям для мужчин и женщин
expenses_by_gender_mcc = transactions_merged[transactions_merged['amount'] < 0].groupby(['gender', 'mcc_code'])['amount'].sum().reset_index()

# 2. Топ-10 категорий с наибольшими относительными модулями разности
# Создаем отдельные датафреймы для мужчин и женщин
male_expenses = expenses_by_gender_mcc[expenses_by_gender_mcc['gender'] == 0][['mcc_code', 'amount']].rename(columns={'amount': 'male_amount'})
female_expenses = expenses_by_gender_mcc[expenses_by_gender_mcc['gender'] == 1][['mcc_code', 'amount']].rename(columns={'amount': 'female_amount'})

# Объединяем
expenses_comparison = pd.merge(male_expenses, female_expenses, on='mcc_code', how='inner')

# Вычисляем относительную разницу
expenses_comparison['relative_diff'] = abs(expenses_comparison['male_amount'] - expenses_comparison['female_amount']) / ((abs(expenses_comparison['male_amount']) + abs(expenses_comparison['female_amount'])) / 2)

# Топ-10 по относительной разнице
top_10_relative_diff = expenses_comparison.nlargest(10, 'relative_diff')

# Добавляем описания mcc-кодов
top_10_with_descriptions = top_10_relative_diff.merge(tr_mcc_codes, on='mcc_code', how='left')

print("Топ-10 категорий с наибольшими относительными разницами в тратах:")
print(top_10_with_descriptions[['mcc_code', 'mcc_description', 'relative_diff']])
````

````python
# Найдите категорию транзакций, в которой совершается больше всего трат в ночное время
# (промежуток с 00:00 по 06:00).
# Выведите название категории и среднюю сумму трат в ней.

import pandas as pd

# Загрузка данных
transactions = pd.read_csv('transactions.csv', delimiter=',')
gender_train = pd.read_csv('gender_train.csv', delimiter=',')
tr_mcc_codes = pd.read_csv('tr_mcc_codes.csv', delimiter=';')
tr_types = pd.read_csv('tr_types.csv', delimiter=';')

# Объединение таблиц
transactions = pd.merge(transactions, gender_train, on='customer_id', how='left')
transactions = pd.merge(transactions, tr_mcc_codes, on='mcc_code', how='inner')
transactions = pd.merge(transactions, tr_types, on='tr_type', how='inner')


# 1. Выделяем час из tr_datetime
# Формат: '15 01:40:52' -> день и время, берем час
def extract_hour(datetime_str):
    try:
        # Разделяем по пробелу и двоеточию
        time_part = datetime_str.split()[1]  # берем часть после пробела
        hour = int(time_part.split(':')[0])  # берем часы
        return hour
    except:
        return None

transactions['tr_hour'] = transactions['tr_datetime'].apply(extract_hour)

# 2. Фильтруем траты в ночное время (только отрицательные суммы, 0-6 часов)
night_spends = transactions[
    (transactions['amount'] < 0) &
    (transactions['tr_hour'] >= 0) &
    (transactions['tr_hour'] < 6)
]

print(f"Всего трат в ночное время: {len(night_spends)}")

if len(night_spends) == 0:
    print("Нет транзакций в ночное время")
else:
    # 3. Группируем по категориям (mcc_description) и считаем статистику
    category_stats = night_spends.groupby('mcc_description').agg(
        total_spends=('amount', 'count'),  # количество трат
        total_amount=('amount', 'sum'),    # общая сумма трат
        avg_amount=('amount', 'mean')      # средняя сумма трат
    ).reset_index()

    # Сортируем по количеству трат (убывание)
    category_stats = category_stats.sort_values('total_spends', ascending=False)

    # 4. Находим категорию с наибольшим количеством трат
    top_category = category_stats.iloc[0]

    # 5. Выводим результат
    print("\n" + "="*70)
    print("РЕЗУЛЬТАТ ДОПОЛНИТЕЛЬНОГО ЗАДАНИЯ:")
    print("="*70)
    print(f"Категория с наибольшим количеством трат в ночное время:")
    print(f"  Название: {top_category['mcc_description']}")
    print(f"  Количество трат: {top_category['total_spends']}")
    print(f"  Общая сумма трат: {abs(top_category['total_amount']):.2f}")
    print(f"  Средняя сумма трат: {abs(top_category['avg_amount']):.2f}")
    print("="*70)

    # Дополнительно: покажем топ-5 категорий
    print("\nТоп-5 категорий по количеству трат в ночное время:")
    print("-"*70)
    for i, (_, row) in enumerate(category_stats.head(5).iterrows(), 1):
        print(f"{i}. {row['mcc_description'][:50]}...")
        print(f"   Количество: {row['total_spends']}, Средняя сумма: {abs(row['avg_amount']):.2f}")
````
